Df describe shows that we have columns with high values & columns with low values. Hence, we need to scale it for our model to learn better.

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

ds_x = df2.drop('y', axis = 'columns')
y_balanced = df2['y']
y_balanced.value_counts()
```

6. Fit into model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

dt = DecisionTreeClassifier()
lg = LogisticRegression()
knn = KNeighborsClassifier()

x_train_bal, x_test_bal, y_train_bal, y_test_bal = train_test_split(x_balanced,y_balanced, test_size = 0.3, random_state = 11)

for i in [dt, lg, knn]:
  i.fit(x_train_bal, y_train_bal)
  pred = i.predict(x_test_bal)
  test_score = accuracy_score(y_test_bal, pred)
  train_score = accuracy_score(y_train_bal,i.predict(x_train_bal))
  if abs(train_score - test_score) <= 0.1:
    print(i)
    print('Accuracy score for train data', accuracy_score(y_test_bal, pred))
    print('Accuracy score for test data', accuracy_score(y_train_bal, i.predict(x_train_bal)))
    print(classification_report(y_test_bal, pred))
    print(confusion_matrix(y_test_bal, pred))
    print('-----------------------------------------')

  else:
    print('nothing happened')

## condition is changed from 0.01 to 0.1 to allow for detection in difference, else it would not detect.
```

```python
## even with treated balanced dataset, the model shows too high accuracy. Need to measure using F1-Score & AUC-ROC curve to rule out overfitting
```

8. Measure with F1-Score & AUC-ROC curve besides accuracy

```python
## since it is multiclass, need to use ovr
## we chose Decision Tree & Logistic Regression


from sklearn.metrics import f1_score, roc_auc_score
from sklearn.preprocessing import label_binarize

# Train and evaluate models
for model in [dt, lg, knn]:
    model.fit(x_train_bal, y_train_bal)
    pred = model.predict(x_test_bal)

    # Calculate F1 score
    f1 = f1_score(y_test_bal, pred, average='weighted')
    print("F1 Score:", f1)

    # Convert the true labels to binary format
    y_test_bin = label_binarize(y_test_bal, classes=[0, 1, 2])

    # Calculate AUC-ROC for each class using OvR strategy
    auc_roc = roc_auc_score(y_test_bin, model.predict_proba(x_test_bal), multi_class='ovr')
    print("AUC-ROC:", auc_roc)
```

EXPLAIN THE OUTLIER PART

```python
# Specify the columns to remove outliers from
columns = ['onPageTime', 'nextTime', 'user_active_days', 'total_visit_counts', 'total_unique_session',
           'time_spent_per_session', 'timeOnPage']

# Define the IQR multiplier for outlier detection
iqr_multiplier = 1.5

# Loop over the columns and remove outliers using IQR method
for column in columns:
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - iqr_multiplier * iqr
    upper_bound = q3 + iqr_multiplier * iqr
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
# Create box plots for the specified columns
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(12, 8))
axes = axes.flatten()

for i, column in enumerate(columns):
    ax = axes[i]
    ax.boxplot(df[column])
    ax.set_title(column)

plt.tight_layout()
plt.show()
```
