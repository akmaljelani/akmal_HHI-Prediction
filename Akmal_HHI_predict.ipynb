```python
def process_data():

  df.dropna(inplace=True)

  ##choosing variables

  df2 = df[['onPageTime', 'nextTime', 'user_active_days',
          'total_visit_counts', 'total_unique_session', 'total_time_spent', 'time_spent_per_session', 'timeOnPage']]

  # #label encoder
  # le = LabelEncoder()
  # df2['isMobile'] = le.fit_transform(df2['isMobile'])

  # # Apply Box-Cox transformation to the skewed variable
  # transformed_variable, lambda_ = stats.boxcox(df2['timeOnPage'])

  # # Update the dataframe with the transformed variable
  # df2['timeOnPage'] = transformed_variable

  # one hot encoder

  # columns_one_hot = ['mobileDeviceBranding', 'operatingSystem', 'country','city']

  # Perform one-hot encoding
  # df_encoded = pd.get_dummies(df2, columns=columns_one_hot)
  ## to handle unknown one hot encoding variables (city, country,etc)
  # df_encoded = pd.get_dummies(df2, columns=columns_one_hot, dummy_na=True)


  # standard scalar
  sc = StandardScaler()

  dataset = sc.fit_transform(df2)

  x = pd.DataFrame(dataset, columns=df2.columns)


  return x
```
